```python
# Reload the original data
nn_tuning_log_df = log_df

# Correctly convert the 'hyperparameters' column from a string to a dictionary
nn_tuning_log_df['hyperparameters'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: eval(x))

# Extract the relevant hyperparameters based on the provided example structure
nn_tuning_log_df['n_layers'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('n_layers', None))
nn_tuning_log_df['layer_sizes'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('layer_sizes', None))
nn_tuning_log_df['activations'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('activations', None))  # Note: the key is 'activations' not 'activation'
nn_tuning_log_df['optimizer'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('optimizer', None))
nn_tuning_log_df['loss'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('loss', None))
nn_tuning_log_df['dropout_rate'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('dropout_rate', None))
nn_tuning_log_df['learning_rate'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('learning_rate', None))
nn_tuning_log_df['momentum'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('momentum', None))
nn_tuning_log_df['epochs'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('epochs', None))
nn_tuning_log_df['batch_size'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('batch_size', None))
nn_tuning_log_df['batch_normalization'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('batch_normalization', None))
nn_tuning_log_df['weight_initializer'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('weight_initializer', None))
nn_tuning_log_df['input_shape'] = nn_tuning_log_df['hyperparameters'].apply(lambda x: x.get('input_shape', None))

# Convert 'activations' to a comma-separated string
nn_tuning_log_df['activations'] = nn_tuning_log_df['activations'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# Drop the original 'hyperparameters' column
nn_tuning_log_df.drop('hyperparameters', axis=1, inplace=True)

# Display the first few rows of the cleaned data to verify if all information is correctly extracted
nn_tuning_log_df.head()
```

```python
# Create the line plot without the "Percentage of Data"
fig, ax1 = plt.subplots(figsize=(14, 10))

# Plot R^2 score with epochs on the x-axis
ax1.plot(nn_tuning_log_df['epochs'], nn_tuning_log_df['score'], label='R^2 Score', color='b', marker='o')
ax1.set_xlabel('Epochs', fontsize=14)
ax1.set_ylabel('R^2 Score', fontsize=14, color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_title('Line Plot of Neural Network Tuning Metrics', fontsize=16)

# Create a second y-axis to plot batch size
ax2 = ax1.twinx()
ax2.plot(nn_tuning_log_df['epochs'], nn_tuning_log_df['batch_size'], label='Batch Size', color='g', marker='x')
ax2.set_ylabel('Batch Size', fontsize=14, color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Add a unified legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

plt.show()
```

```python
# Create the scatter plot
fig, ax1 = plt.subplots(figsize=(14, 8))

# Normalize the learning rate for better visualization
normalized_r_score = (nn_tuning_log_df['score'] + -1*nn_tuning_log_df['score'].min())
scaled_sizes = normalized_r_score * 100  # Scale sizes

# Scatter Epochs with R^2 score on the x-axis, size of the marker represents Scaled Learning Rate
ax1.scatter(nn_tuning_log_df['batch_size'], nn_tuning_log_df['epochs'], s=scaled_sizes, color='b', alpha=0.6)
ax1.set_xlabel('R^2 Score', fontsize=14)
ax1.set_ylabel('Epochs', fontsize=14)
ax1.tick_params(axis='y',)
ax1.set_title('Epochs vs R^2 Score (Scatter Plot with Batch Size as Size)', fontsize=16)
ax1.grid(True)  # Add grid lines for better visibility

# Add a unified legend
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines, labels, loc='upper right')

plt.tight_layout()
plt.show()


```
